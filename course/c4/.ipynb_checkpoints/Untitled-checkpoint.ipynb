```python
import numpy as np
import sympy as sp


from sympy import init_printing
init_printing(use_latex=True)
```

```python
x, nu, t = sp.symbols('x nu t')
phi = (sp.exp(-(-4 * t + x)**2 / (4 * nu * (t + 1))) +
       sp.exp(-(x - 4 * t - 2 * sp.pi)**2 / (4 * nu * (t + 1))))
phi
```

```python
phiprime = phi.diff(x)
phiprime
```
